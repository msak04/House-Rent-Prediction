```python
from sklearn.model_selection import train_test_split

X = dataFig.drop(['Price'], axis=1)
y = dataFig.Price

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler

# fit scaler on training data
norm = MinMaxScaler().fit(X_train)

# transform training data
X_train_norm = norm.transform(X_train)

# transform testing dataabs
X_test_norm = norm.transform(X_test)
```

```python
from sklearn.preprocessing import StandardScaler

# copy of datasets
X_train_stand = X_train.copy()
X_test_stand = X_test.copy()

# numerical features
num_cols = ['Type', 'Beds', 'Bath', 'Size', 'Educational Institute','Restaurants', 'Medical Service', 'Parks']

# apply standardization on numerical features
for i in num_cols:
    
    # fit on training data column
    scale = StandardScaler().fit(X_train_stand[[i]])
    
    # transform the training data column
    X_train_stand[i] = scale.transform(X_train_stand[[i]])
    
    # transform the testing data column
    X_test_stand[i] = scale.transform(X_test_stand[[i]])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score,mean_absolute_error,mean_absolute_error
from sklearn.metrics import accuracy_score
  
 # create regressor object
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)
  
# fit the regressor with x and y data
#regressor.fit(X_train, y_train)

#Y_pred = regressor.predict(X_train)
#cm1 = confusion_matrix( y_test, pred)
#print("Confusion Matrix : ", cm1)
#print ("Accuracy : ", accuracy_score(y_test, Y_pred))
rmseR = []
mseR=[]
maeR=[]


# raw, normalized and standardized training and testing data
trainX = [X_train, X_train_norm, X_train_stand]
testX = [X_test, X_test_norm, X_test_stand]

# model fitting and measuring RMSE
for i in range(len(trainX)):
    
    # fit
    regressor.fit(trainX[i],y_train)
    # predict
    pred = regressor.predict(testX[i])
    # RMSE
    rmseR.append(np.sqrt(mean_squared_error(y_test,pred)))
    mseR.append(mean_squared_error(y_test,pred))
    maeR.append(mean_absolute_error(y_test,pred))

# visualizing the result    
df_RM = pd.DataFrame({'Root Mean Square Error(RMSE)':rmseR},index=['Original','Normalized','Standardized'])
print(df_RM)
df_RME = pd.DataFrame({'Mean Squared Error (MSE)':mseR},index=['Original','Normalized','Standardized'])
print(df_RME)
df_RMA = pd.DataFrame({'Mean Absolute Error (MAE)':maeR},index=['Original','Normalized','Standardized'])
print(df_RMA)
```

```python
# training a Decision Tree model
from sklearn.tree import DecisionTreeRegressor
# measuring RMSE score
from sklearn.metrics import mean_squared_error


# Decision tree
dt = DecisionTreeRegressor(max_depth=10,random_state=27)

rmseD = []
mseD=[]
maeD=[]

# raw, normalized and standardized training and testing data
trainX = [X_train,X_train_norm,X_train_stand]
testX = [X_test,X_test_norm,X_test_stand]

# model fitting and measuring RMSE
for i in range(len(trainX)):
    
    # fit
    dt.fit(trainX[i],y_train)
    # predict
    pred = dt.predict(testX[i])
    # RMSE
    rmseD.append(np.sqrt(mean_squared_error(y_test,pred)))
    mseD.append(mean_squared_error(y_test,pred))
    maeD.append(mean_absolute_error(y_test,pred))

# visualizing the result    
df_dt = pd.DataFrame({'Root Mean Square Error(RMSE)':rmseD},index=['Original','Normalized','Standardized'])
print(df_dt)
df_dt1 = pd.DataFrame({'Mean Squared Error (MSE)':mseD},index=['Original','Normalized','Standardized'])
print(df_dt1)
df_dt2 = pd.DataFrame({'Mean Absolute Error (MAE)':maeD},index=['Original','Normalized','Standardized'])
print(df_dt2)
```
